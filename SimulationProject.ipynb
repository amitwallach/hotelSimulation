  <p>לכן, ניתן להסיק שהמודל של ההתפלגות המעריכית מתאים לתיאור התנהגות זמני הצ'ק-אין והצ'ק-אאוט של המלון, וכי פרמטר הקצב <math>λ</math> יוכל לשמש אותנו בהמשך לצורך סימולציות וחיזוי דפוסי ההגעה והעזיבה של הלקוחות.</p>
  <p>על מנת לאשש את הטענה כי ההתפלגות היא אכן מעריכית, נבצע <strong>מבחן KS</strong>.</p>

<div style="direction:rtl" dir="rtl">
  <h2><strong>מציאת אומד נראות מקסימאלי (MLE) וביצוע מבחן KS</strong></h2>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Function to compute the empirical CDF
def compute_empirical_cdf(data):

    data_sorted = np.sort(data)
    empirical_cdf_values = np.arange(1, len(data) + 1) / len(data)
    return data_sorted, empirical_cdf_values

# Function for the Kolmogorov-Smirnov test
def ks_test(data, theoretical_cdf, *params):

    data_sorted, empirical_cdf_values = compute_empirical_cdf(data)
    theoretical_cdf_values = theoretical_cdf(data_sorted, *params)
    d_statistic = np.max(np.abs(empirical_cdf_values - theoretical_cdf_values))
    return d_statistic, data_sorted, empirical_cdf_values, theoretical_cdf_values

# Define the theoretical CDF for the exponential distribution
def exp_cdf(x, lambda_):

    return 1 - np.exp(-lambda_ * x)

# Function to process a dataset and perform the KS test
def process_and_ks_test(file_path, label):

    # Read the data from the Excel file
    data = pd.read_excel(file_path).values.flatten()
    data = data[~np.isnan(data)]  # Remove NaN values

    # Estimate lambda (MLE)
    lambda_mle = 1 / np.mean(data)
    print(f"lambda mle: {lambda_mle}")

    # Perform KS test
    d_statistic, data_sorted, empirical_cdf_values, theoretical_cdf_values = ks_test(data, exp_cdf, lambda_mle)

    # Approximate the p-value
    n = len(data)
    p_value = np.exp(-2 * (d_statistic**2) * n)

    # Print results
    print(f"{label} Results:")
    print(f"KS Statistic: {d_statistic}")
    print(f"P-value: {p_value}")
    if p_value > 0.05:
        print("Fail to reject the null hypothesis (H0). The data follows the exponential distribution.")
    else:
        print("Reject the null hypothesis (H0). The data does not follow the exponential distribution.")

    # Visualization
    plt.figure(figsize=(8, 6))
    plt.step(data_sorted, empirical_cdf_values, where='post', label="Empirical CDF", color='blue')
    plt.plot(data_sorted, theoretical_cdf_values, label="Theoretical CDF (Exponential)", color='red', linestyle='--')
    plt.title(f"Empirical vs. Theoretical CDF for {label}")
    plt.xlabel("Data")
    plt.ylabel("CDF")
    plt.legend()
    plt.grid()
    plt.show()

# Path to Excel file
checkin_file = '/content/CheckIn.xlsx'

# Process and test for Check-In data
process_and_ks_test(checkin_file, "Check-In")


def compute_empirical_cdf(data):

    data_sorted = np.sort(data)
    empirical_cdf_values = np.arange(1, len(data) + 1) / len(data)
    return data_sorted, empirical_cdf_values

# Function for the Kolmogorov-Smirnov test
def ks_test(data, theoretical_cdf, *params):

    data_sorted, empirical_cdf_values = compute_empirical_cdf(data)
    theoretical_cdf_values = theoretical_cdf(data_sorted, *params)
    d_statistic = np.max(np.abs(empirical_cdf_values - theoretical_cdf_values))
    return d_statistic, data_sorted, empirical_cdf_values, theoretical_cdf_values

# Define the theoretical CDF for the exponential distribution
def exp_cdf(x, lambda_):

    return 1 - np.exp(-lambda_ * x)

# Function to process a dataset and perform the KS test
def process_and_ks_test(file_path, label):

    # Read the data from the Excel file
    data = pd.read_excel(file_path).values.flatten()
    data = data[~np.isnan(data)]  # Remove NaN values

    # Estimate lambda (MLE)
    lambda_mle = 1 / np.mean(data)
    print(f"lambda mle: {lambda_mle}")

    # Perform KS test
    d_statistic, data_sorted, empirical_cdf_values, theoretical_cdf_values = ks_test(data, exp_cdf, lambda_mle)

    # Approximate the p-value
    n = len(data)
    p_value = np.exp(-2 * (d_statistic**2) * n)

    # Print results
    print(f"{label} Results:")
    print(f"KS Statistic: {d_statistic}")
    print(f"P-value: {p_value}")
    if p_value > 0.05:
        print("Fail to reject the null hypothesis (H0). The data follows the exponential distribution.")
    else:
        print("Reject the null hypothesis (H0). The data does not follow the exponential distribution.")

    # Visualization
    plt.figure(figsize=(8, 6))
    plt.step(data_sorted, empirical_cdf_values, where='post', label="Empirical CDF", color='blue')
    plt.plot(data_sorted, theoretical_cdf_values, label="Theoretical CDF (Exponential)", color='red', linestyle='--')
    plt.title(f"Empirical vs. Theoretical CDF for {label}")
    plt.xlabel("Data")
    plt.ylabel("CDF")
    plt.legend()
    plt.grid()
    plt.show()

# Path to Excel file
checkout_file = '/content/Checkout.xlsx'

# Process and test for Check-Out data
process_and_ks_test(checkout_file, "Check-Out")
```

  <h2><strong>זמן שהייה בבריכה - אלגוריתם טרנספורם הופכי </strong></h2>
  <p><strong>הפונקציה מוגדרת כך:</strong></p>
  \[
  f(x) =
  \begin{cases}
  \frac{x}{6}, & 1 \leq x < 2 \\
  \frac{x}{5} + \frac{1}{8}, & 2 \leq x < 3 \\
  \frac{1}{8}, & 3 \leq x < 4
  \end{cases}
  \]


  <h3><strong>פונקציית ההצטברות (CDF)</strong></h3>
  \[
  F(x) =
  \begin{cases}
  \frac{x^2 - 1}{12}, & 1 \leq x < 2 \\
  \frac{1}{4} + \frac{x^2 - 4}{10} + \frac{x - 2}{8}, & 2 \leq x < 3 \\
  \frac{7}{8} + \frac{x - 3}{8}, & 3 \leq x < 4
  \end{cases}
  \]

  <h3><strong>תחום 1 (\( 1 \leq x < 2 \)):</strong></h3>
  \[
  \int_1^x \frac{t}{6} dt = \frac{t^2}{12} \Bigg|_1^x = \frac{x^2}{12} - \frac{1}{12} = \frac{x^2 - 1}{12}
  \]
  <p><strong>בסיום התחום הראשון:</strong></p>
  \[
  F(2) = \frac{2^2 - 1}{12} = \frac{3}{12} = \frac{1}{4}
  \]

  <h3><strong>תחום 2 (\( 2 \leq x < 3 \)):</strong></h3>
  \[
  \int_2^x \left( \frac{t}{5} + \frac{1}{8} \right) dt = \int_2^x \frac{t}{5} dt + \int_2^x \frac{1}{8} dt
  \]
  <p><strong>חישוב האינטגרל הראשון:</strong></p>
  \[
  \int_2^x \frac{t}{5} dt = \frac{t^2}{10} \Bigg|_2^x = \frac{x^2}{10} - \frac{4}{10}
  \]
  <p><strong>חישוב האינטגרל השני:</strong></p>
  \[
  \int_2^x \frac{1}{8} dt = \frac{t}{8} \Bigg|_2^x = \frac{x}{8} - \frac{2}{8}
  \]
  <p><strong>סיכום התחום השני:</strong></p>
  \[
  F(x) = \frac{1}{4} + \left( \frac{x^2}{10} - \frac{4}{10} \right) + \left( \frac{x}{8} - \frac{2}{8} \right)
  \]
  \[
  F(x) = \frac{x^2}{10} + \frac{x}{8} - 0.4
  \]
  <p><strong>בסיום התחום השני:</strong></p>
  \[
  F(3) = \frac{9}{10} + \frac{3}{8} - 0.4 = 0.875
  \]

  <h3><strong>תחום 3 (\( 3 \leq x < 4 \)):</strong></h3>
  \[
  \int_3^x \frac{1}{8} dt = \frac{t}{8} \Bigg|_3^x = \frac{x}{8} - \frac{3}{8}
  \]
  <p><strong>סיכום התחום השלישי:</strong></p>
  \[
  F(x) = 0.875 + \left( \frac{x}{8} - \frac{3}{8} \right) = \frac{x}{8} + 0.875 - \frac{3}{8}
  \]
  \[
  F(x) = \frac{x}{8} + 0.875
  \]

  <h2><strong>חישוב הפונקציה ההפוכה \( F(x) = u \)</strong></h2>
  <p>נמצא את הפונקציה ההפוכה עבור כל תחום של \( x \):</p>

  <h3><strong>תחום 1: \( 1 \leq x < 2 \), \( 0 < u \leq 0.25 \)</strong></h3>
  \[
  \frac{x^2 - 1}{12} = u \implies x^2 - 1 = 12u \implies x^2 = 12u + 1 \implies x = \sqrt{12u + 1}
  \]

  <h3><strong>תחום 2: \( 2 \leq x < 3 \), \( 0.25 < u \leq 0.875 \)</strong></h3>
  \[
  \frac{x^2}{10} + \frac{x}{8} - 0.4 = u
  \]
  <p>נכפיל ב-40:</p>
  \[
  4x^2 + 5x - (16 + 40u) = 0
  \]
  <p><strong>נוסחת השורשים:</strong></p>
  \[
  x = \frac{-5 \pm \sqrt{281 + 640u}}{8}
  \]

  <h3><strong>תחום 3: \( 3 \leq x < 4 \), \( 0.875 < u \leq 1 \)</strong></h3>
  \[
  \frac{x - 3}{8} + 0.875 = u \implies x = 8u - 4
  \]

  <h3><strong>פונקציית הדגימה:</strong></h3>
  <p>דגום \( u \) מתוך \( U(0,1) \)</p>
  <ol>
    <li>אם \( 0 < u \leq 0.25 \): \[ x = \sqrt{12u + 1} \]</li>
    <li>אם \( 0.25 < u \leq 0.875 \): \[ x = \frac{-5 + \sqrt{281 + 640u}}{8} \]</li>
    <li>אם \( 0.875 < u \leq 1 \): \[ x = 8u - 4 \]</li>
  </ol>

```python
import heapq
import random
from fontTools.merge.util import current_time

#-------------------------------------------------------------------------------Customer---------------------------------------------------------------------------------------------------------------
class Customer:
    countCustomers = 0 # Tracks the total number of customers created.
    def __init__(self, groupType, bookingId):
        Customer.countCustomers += 1
        self.customerId = Customer.countCustomers
        self.bookingID = bookingId
        self.expenses = 0
        self.rank = 10
        self.groupType = groupType
        self.last_activity=None
        self.daily_activities, self.maxWaitingTime = self.initializeGroupAttributes()  # Initializes attributes based on group type.

    def initializeGroupAttributes(self):
        """
        Initialize the group's daily activities and maximum waiting time based on the group type.
        """
        groupSettings = {
            "family": ({"pool": [False, 0], "bar": [False, 0]}, 10),
            "couple": ({"spa": [False, 0], "pool": [False, 0]}, 15),
            "individual": ({"spa": [False, 0], "bar": [False, 0], "pool": [False, 0]}, 20)
        }
        return groupSettings.get(self.groupType, ({}, 0))

    def update_rank(self, rank_decrease):
        self.rank = max(self.rank - rank_decrease, 0)

    def add_expense(self, expense):
        self.expenses += expense

    def get_expenses(self):
        return self.expenses

    def checkGroupType(self):
        return self.groupType

    def updateBarExpenses(self):
        """
        Simulate a bar expense for the customer.
        """
        # Define expenses
        drink_expenses = {
            "family": [3],  # Families only have Coffee or Juice ($3)
            "couple": [3, 10, 15],  # Couples have access to all drinks
            "individual": [3, 10, 15]  # Individuals have access to all drinks
        }

        food_expenses = [10, 12, 3, 15]  # Food is the same for all groups

        # Randomize choice
        u1, u2 = np.random.uniform(), np.random.uniform()
        if u1 < 0.5:  # 50% chance for drinks
            expense = self.choose_expense(drink_expenses[self.groupType], u2)
        else:  # 50% chance for food
            expense = self.choose_expense(food_expenses, u2)
        self.add_expense(expense)

    def choose_expense(self, expense_options, probability):
        """
        Choose an expense from the given options based on uniform probability.
        """
        num_options = len(expense_options)
        for i in range(num_options):  # Iterates through expense options to find the one matching the probability.
            if probability < (i + 1) / num_options:
                return expense_options[i]
        return expense_options[-1]

    def __str__(self):
        return (f"Customer ID: {self.customerId}, Booking ID: {self.bookingID}, Group Type: {self.groupType}, "
                f"Rank: {self.rank}, Expenses: {self.expenses}")
#-------------------------------------------------------------------------------Booking--------------------------------------------------------------------------------------------------------------------------
class Booking:
    totalBookings = 0  # Static variable to track the total number of bookings
    def __init__(self, arrivalTime):
        Booking.totalBookings += 1
        self.bookingId = Booking.totalBookings
        self.customers = []  # Changed to lowercase for consistency
        self.numCustomers = sampleCustomerCount()
        self.stayDuration = sampleStayDuration()
        self.arrivalTime = arrivalTime
        self.groupType = self.checkGroupType()
        self.breakfastArrivalTime = 0
        self.hasCheckedOut = False
        for i in range(self.numCustomers):
            self.customers.append(Customer(self.groupType, self.bookingId))

    def __lt__(self, other):
      """
      Compare bookings based on arrival time.
      """
      return self.arrivalTime < other.arrivalTime

    def resetDailyActivities(self):
      """
      Reset the daily activities of all customers in the booking.
      """
      for customer in self.customers:
          customer.lastActivity = None  # Reset last activity for the customer.
          customer.daily_activities, customer.maxWaitingTime = customer.initializeGroupAttributes()

    def getGroupSize(self):
        return self.numCustomers # Return the number of customers in the booking.

    def getDurationtime(self):
        return self.stayDuration

    def checkGroupType(self):
        """
        Categorize the group based on its size.
        """
        if self.numCustomers >= 3: # Groups with 3 or more customers are families.
            return "family"
        elif self.numCustomers == 2: # Groups with 2 customers are couples.
            return "couple"
        return "individual" # Single customers are categorized as individuals.

    def isLastDay(self, currentTime):
        """
        Check if it's the last day of the group's stay.
        """
        arrivalDay = self.arrivalTime // 1440+1
        lastDay = arrivalDay + self.stayDuration
        currentDay = currentTime
        return currentDay == lastDay

    def isFirstDay(self, currentTime):
        """
        Check if it's the first day of the group's stay.
        """
        arrivalDay = self.arrivalTime // 1440 + 1
        currentDay = currentTime
        return currentDay == arrivalDay

    def bookingRank(self):
        """
        Calculate the average rank of all customers in the booking.
        """
        rankSum = 0
        rankSum = sum(customer.rank for customer in self.customers)
        return rankSum / self.numCustomers

    def bookingCost(self, roomType):
        """
        Calculate the total cost of the booking.
        """
        totalCost = 0
        if roomType == "Suite":
            totalCost += self.stayDuration * 370
        else:
            totalCost += self.stayDuration * 250

        for customer in self.customers:
            totalCost += customer.get_expenses()
        return totalCost
#--------------------------------------------------------------------------------Room------------------------------------------------------------------------------------------------------------------
class Room:
    totalRooms = 0  # Static variable to track the total number of rooms created
    def __init__(self, roomType):
        Room.totalRooms += 1 # Increment the total rooms counter for each new room.
        self.roomId = Room.totalRooms
        self.roomType = roomType
        self.currentBooking = None  # The guest group currently occupying the room
        self.available = True  # Indicates if the room is free

    def isAvailable(self):
        """
        Check if the room is free.
        """
        return self.available # Returns True if the room is not occupied.

    def bookRoom(self, booking):
        """
        Assign a guest group to the room and mark it as occupied.
        """
        self.currentBooking = booking
        self.available = False

    def getCurrentBooking(self):
        """
        Return the current guest group occupying the room.
        """
        return self.currentBooking

    def change_to_available_room(self):
        """
        Marks the room as available and releases the current booking.
        """
        if self.currentBooking:
            self.currentBooking = None
            self.available = True
#-------------------------------------------------------------------------------Facilities-----------------------------------------------------------------------------------------------------------
class Bar:
    def __init__(self, numberOfBarmens):
        self.barQueue = [] # Queue of customers waiting for service at the bar.
        self.num_barmen = numberOfBarmens # Number of barmen available.
        self.barmen_states = [False] * numberOfBarmens # Barmen availability states (False = available, True = busy)

    def findAvailableBarman(self):
        """
        Find an available barman. return: Index of an available barman if found, otherwise Non
        """
        for i, is_busy in enumerate(self.barmen_states):
            if not is_busy:
                return i
        return None

    def setBarmanStatus(self, barman_index, status):
        """
        Update the status of a barman. status: True if the barman is busy, False if available.
        """
        self.barmen_states[barman_index] = status

class Facilities:
    def __init__(self, activityType, maxCapacity):
        self.activityType = activityType
        self.maxCapacity = maxCapacity
        self.currentOccupancy = 0
        self.activityQueue = []  # Queue for guests waiting to participate
        self.groupListActivity = []  # List of groups currently in the activity

    def checkActivityAvailabilityforagust(self):
        requiredCapacity = 1 # Capacity needed for one guest.
        return self.currentOccupancy + requiredCapacity <= self.maxCapacity

    def checkIfQueueIsNotEmpty(self):
        """
        Check if there are customers in the activity queue.
        """
        return len(self.activityQueue) > 0

    def addCustomerToActivity(self, num_of_guests):
        """
        Add a group of customers to the activity if there is enough capacity.
        """
        requiredCapacity = num_of_guests
        if self.currentOccupancy + requiredCapacity <= self.maxCapacity:
            self.currentOccupancy += num_of_guests # Increase the occupancy by the group size.
            return True
        return False

    def endCustomerActivity(self, num_of_guests):
        """
        Remove a group of customers from the activity, freeing up capacity.
        """
        self.currentOccupancy -= num_of_guests

    def addToActivityQueue(self, customer):
        """
        Add a customer to the activity queue if they are not already in it.
        """
        if customer not in self.activityQueue:
            self.activityQueue.append(customer)

class Spa(Facilities):
    def __init__(self):
        super().__init__(activityType="Spa", maxCapacity=30)
        self.spaQueue = []

class Pool(Facilities):
    def __init__(self):
        super().__init__(activityType="Pool", maxCapacity=50)
        self.poolQueue = []

class Table:
    def __init__(self):
        self.is_occupied = False # Tracks whether the table is currently occupied.

    def set_status(self, status):
        """
        Update the table's occupancy status. True if the table is occupied, False if available.
        """
        self.is_occupied = status
#-------------------------------------------------------------------------------Hotel-----------------------------------------------------------------------------------------------------------
class Hotel:
    def __init__(self, numberOfTables, numberOfBarmens):
        self.check_in_queue = []
        self.check_out_queue = []
        self.lobby_queue = []
        self.num_servers = 2 # Number of servers available
        self.server_states = [False] * self.num_servers
        self.family_rooms = [Room("family") for _ in range(30)]
        self.triple_rooms = [Room("triple") for _ in range(40)]
        self.couple_rooms = [Room("couple") for _ in range(30)]
        self.suite_rooms = [Room("suite") for _ in range(10)]
        self.free_rooms_count = 110 # Total number of free rooms
        self.tables = [Table() for _ in range(numberOfTables)]
        self.breakfast_queue = []
        self.daily_rank = 7 # Initial hotel rank
        self.hotelLambada = 9.8 # Hotel Lambda value for optimization
        self.hotelRatings  = [7]
        self.facilities = self.initializeFacilities() # Facilities like pool and spa
        self.bar = Bar(numberOfBarmens) # Initialize the bar
        self.breakfast_list = [] # List of breakfast bookings
        self.dailyactivity_list = [] # Daily activity list

        self.dailyHotelRanking = [7]
        self.dailyBreakfastWaitingTime = [0]
        self.bookingsBreakfastWaitingTime = [] #Init every day
        self.dailyExpenses = []
        self.hotelIncomes = []
        self.bar_arrivals = 0  # Total arrivals at the bar
        self.bar_reneging_events = 0  # Total abandonment events at the bar
        self.daily_bar_abandonment_rates = []  # Daily abandonment rates at the bar

    def calculate_daily_breakfast_waiting_time(self):
        """
        Calculate the average daily waiting time for breakfast and update the dailyBreakfastWaitingTime array.
        """
        if not self.bookingsBreakfastWaitingTime:  # If there are no bookings with breakfast waiting times
            self.dailyBreakfastWaitingTime.append(0)
        else:
            daily_average = sum(self.bookingsBreakfastWaitingTime) / len(self.bookingsBreakfastWaitingTime)
            self.dailyBreakfastWaitingTime.append(daily_average)
        self.bookingsBreakfastWaitingTime=[]

    def update_daily_rank(self):
        """
        Update the daily rank of the hotel based on customer feedback.
        """
        self.daily_rank = sum(self.hotelRatings) / len(self.hotelRatings)
        self.dailyHotelRanking.append(self.daily_rank)

    def update_daily_expenses(self):
        if not self.dailyExpenses:  # If there are no expenses for the day
              self.hotelIncomes.append(0)
        else:
              daily_incomes = sum(self.dailyExpenses)  # Calculate the total daily income
              self.hotelIncomes.append(daily_incomes)  # Append to hotel incomes
        self.dailyExpenses = []  # Reset daily expenses for the next day

    def calculate_daily_bar_abandonment_rate(self):
      """
      Calculate the daily bar abandonment rate based on arrivals and reneging events
      """
      if self.bar_arrivals == 0:
            rate = 0
      else:
            rate = (self.bar_reneging_events / self.bar_arrivals) * 100
      self.daily_bar_abandonment_rates.append(rate)
      self.bar_reneging_events = 0
      self.bar_arrivals = 0


    def check_and_assign_room(self, booking):
        """
        Check if a room is available for the booking, and assign it if possible.
        """
        booking_size = booking.getGroupSize()

        if booking_size >= 4:
            room_list = self.family_rooms
        elif booking_size == 3:
            room_list = self.triple_rooms
        elif booking_size == 2:
            room_list = self.suite_rooms if sampleIsSuite(2) else self.couple_rooms
        elif booking_size == 1:
            room_list = self.suite_rooms if sampleIsSuite(1) else self.couple_rooms

        for room in room_list:
            if room.isAvailable():
                room.bookRoom(booking) # Assign the room to the booking
                return room.roomId
        return None

    def collectBreakfastBookings(self):
      """
      Collects active bookings for breakfast from all room types and stores them in the `breakfast_list`.
      """
      self.breakfast_list = []
      for room_list in [self.family_rooms, self.triple_rooms, self.couple_rooms, self.suite_rooms]:
          for room in room_list:
              booking = room.getCurrentBooking()
              if booking:
                  self.breakfast_list.append(booking)

    def calculateAvailableRoomsByTypeAtMidnight(self, current_date):
      """
      Counts and updates available rooms by type at midnight, returning a dictionary with the totals.
      """

      available_rooms = { "family": 0, "triple": 0, "couple": 0, "suite": 0 }
      for room_list, room_type in [
          (self.family_rooms, "family"),
          (self.triple_rooms, "triple"),
          (self.couple_rooms, "couple"),
          (self.suite_rooms, "suite")
      ]:
          for room in room_list:
              booking = room.getCurrentBooking()
              if booking and booking.isLastDay(current_date):
                  room.change_to_available_room()
                  available_rooms[room_type] += 1
              elif room.isAvailable():
                  available_rooms[room_type] += 1
      total_available = sum(available_rooms.values())
      return available_rooms

    def count_free_rooms(self):
        return sum(
            1 for room_list in [self.family_rooms, self.triple_rooms, self.couple_rooms, self.suite_rooms]
            for room in room_list if room.isAvailable()
        )

    def release_room(self, booking):
        for room_list in [self.family_rooms, self.triple_rooms, self.couple_rooms, self.suite_rooms]:
            for room in room_list:
                if room.getCurrentBooking() == booking:
                    room.change_to_available_room()
                    return

    def find_available_server(self):
        for i, busy in enumerate(self.server_states):
            if not busy:
                return i
        return None

    def assign_to_server(self, server_id ):
        self.server_states[server_id] = True

    def find_available_table(self):
        for index, table in enumerate(self.tables):
            if not table.is_occupied:
                return index
        return None

    def get_daily_rank(self):
        return self.daily_rank

    def calculate_lambda(self,available_rooms):
        alpha = 20
        R_total = 110
        beta_1 = 1.5
        beta_2 = 2
        R_available = available_rooms
        H = self.daily_rank
        R_indicator=self.is_room_available_indicator()

        if R_available == 0:
            self.hotelLambada = 0
        else:
            self.hotelLambada = alpha * ((R_available / R_total) ** beta_1) * ((H / 10) ** beta_2)*R_indicator

        return self.hotelLambada

    def get_hotal_lambda(self):
        return self.hotelLambada

    def is_room_available_indicator(self):
        return 1 if self.count_free_rooms() > 0 else 0

    def initializeFacilities(self):
        """
        Create facilities including the pool and spa.
        """
        return {
            "pool": Pool(),
            "spa": Spa()
        }

    def getFacility(self, facility_name):
        """
        Get a specific facility by name.
        """
        return self.facilities.get(facility_name)

    def resetAllBookingsDailyActivities(self):
        for room_list in [self.family_rooms, self.triple_rooms, self.couple_rooms, self.suite_rooms]:
            for room in room_list:
                if room.currentBooking:
                    room.currentBooking.resetDailyActivities()

    def getNextActivity(self, customer, activities_dict, last_activity):
        if customer.groupType == "family":
            for activity, status in activities_dict.items():
                success, attempts = status
                if last_activity == "pool" and success:
                    if activity != last_activity and attempts < 2:
                        return activity
                elif last_activity == "pool" and not success:
                    if activity != last_activity and attempts < 2:
                        return activity
                elif last_activity == "bar" and success:
                    if activity != last_activity and attempts < 2:
                        return activity
                elif last_activity == "bar" and not success:
                    if activity != last_activity and attempts < 2:
                        return activity

        elif customer.groupType == "couple":
            for activity, status in activities_dict.items():
                success, attempts = status
                if last_activity == "pool" and success:
                    if activity != last_activity and attempts < 2:
                        return activity
                elif last_activity == "pool" and not success:
                    if activity != last_activity and attempts < 2:
                        return activity
                elif last_activity == "spa" and success:
                    if activity != last_activity and attempts < 2:
                        return activity
                elif last_activity == "spa" and not success:
                    if activity != last_activity and attempts < 2:
                        return activity

        elif customer.groupType == "individual":
            for activity, status in activities_dict.items():
                success, attempts = status
                if last_activity == "pool" and success:
                    if activity != last_activity and attempts < 2:
                        return activity
                elif last_activity == "pool" and not success:
                    if activity != last_activity and attempts < 2:
                        return activity
                elif last_activity == "spa" and success:
                    if activity != last_activity and attempts < 2:
                        return activity
                elif last_activity == "spa" and not success:
                    if activity != last_activity and attempts < 2:
                        return activity
                elif last_activity == "bar" and success:
                    if activity != last_activity and attempts < 2:
                        return activity
                elif last_activity == "bar" and not success:
                    if activity != last_activity and attempts < 2:
                        return activity
class Event:
    def __init__(self, time):
        self.time = time

    def __lt__(self, other):
        return self.time < other.time

    def handle(self, simulation):
        raise NotImplementedError("Handle method must be implemented by subclasses")

class OpenHotelEvent(Event):
    """
    Open the hotel to guests at 8am
    """
    def handle(self, simulation):
        # Calculate the time when customer arrivals will end (5pm)
        arrival_end_time = self.time + 9 * 60  # 9 hours after opening time

        # Schedule an event to stop customer arrivals at the calculated end time
        simulation.scheduleEvent(CloseCustomerArrivalEvent(arrival_end_time))

        # Determine the next customer arrival time based on the arrival rate (lambda)
        next_arrival_time = self.time + sampleCustomerArrival(simulation.Hotel.get_hotal_lambda())

        # If the next arrival time is within the arrival window, schedule a check-in arrival event
        if next_arrival_time < arrival_end_time:
            simulation.scheduleEvent(CheckInArrivalEvent(next_arrival_time))

        # Schedule the start of check-in processing at 3pm (7 hours after opening)
        check_in_start_time = self.time + 7 * 60
        for _ in range(2):
              if simulation.Hotel.lobby_queue:
                  booking = heapq.heappop(simulation.Hotel.lobby_queue)
                  simulation.scheduleEvent(CheckInDepartureEvent(check_in_start_time, booking))


class closeHotelDayEvent(Event):
    """
    Handles end-of-day operations and schedules the next day's opening at 00:00.
    """
    def handle(self, simulation):
        # Calculate the current date in days (integer division of time by minutes in a day)
        current_date = self.time // (24 * 60)

        # Perform various end-of-day updates for the hotel
        simulation.Hotel.collectBreakfastBookings()  # Collect bookings for breakfast
        simulation.Hotel.update_daily_expenses()  # Update daily expenses for the hotel
        simulation.Hotel.calculate_daily_bar_abandonment_rate()  # Calculate bar abandonment rates

        # Count the number of free rooms at the end of the day
        available_rooms = simulation.Hotel.count_free_rooms()

        # Reset daily activities and bookings for all guests
        simulation.Hotel.resetAllBookingsDailyActivities()

        # Update hotel rank and calculate breakfast waiting times
        simulation.Hotel.update_daily_rank()
        simulation.Hotel.calculate_daily_breakfast_waiting_time()

        # Predict room availability for the next day at midnight
        available_rooms_nextday = simulation.Hotel.calculateAvailableRoomsByTypeAtMidnight(current_date + 1)

        # Update the lambda value (arrival rate) based on available rooms
        current_lamda = simulation.Hotel.calculate_lambda(available_rooms)

        # Retrieve and update the daily rank
        daily_rank = simulation.Hotel.get_daily_rank()

        # Schedule the event to close the next day at 00:00
        simulation.scheduleEvent(closeHotelDayEvent(self.time + 24 * 60))

        # Reset the daily count for people checking out after breakfast
        simulation.Hotel.count_people_checkout_after_breakfast = 0

class CloseCustomerArrivalEvent(Event):
    """
    Handles the closing of customer arrivals at 5pm and schedules the hotel reopening at 8am the next day.
    """
    def handle(self, simulation):
        # Schedule the hotel reopening event for 8am the following day (15 hours after closing)
        simulation.scheduleEvent(OpenHotelEvent(self.time + 15 * 60))

class CheckInArrivalEvent(Event):
    """
    Handles the arrival of a customer for check-in and adds them to the lobby queue if a room is available.
    """
    def handle(self, simulation):
        # Calculate the next arrival time based on the hotel's arrival rate (lambda)
        next_arrival_time = self.time + sampleCustomerArrival(simulation.Hotel.get_hotal_lambda())

        # Convert the next arrival time to local time within the day (minutes since midnight)
        local_time = next_arrival_time % 1440

        # Define the check-in window (8:00 AM to 5:00 PM)
        check_in_start_time = 8 * 60  # 8:00 AM in minutes
        check_in_end_time = 17 * 60  # 5:00 PM in minutes

        # If the next arrival is outside the check-in window, do nothing
        if not (check_in_start_time <= local_time <= check_in_end_time):
            return

        # Schedule the next check-in arrival event
        simulation.scheduleEvent(CheckInArrivalEvent(next_arrival_time))

        # Create a new booking for the customer
        booking = Booking(self.time)

        # Check if a room is available and assign it; if successful, add to the lobby queue
        if simulation.Hotel.check_and_assign_room(booking):
            heapq.heappush(simulation.Hotel.lobby_queue, booking)

class CheckInDepartureEvent(Event):
    def __init__(self, time, booking=None):
        super().__init__(time)
        self.booking = booking

    def handle(self, simulation):
            # If the lobby queue is empty, return immediately
            if not simulation.Hotel.lobby_queue and not self.booking:
                return

            # If no specific booking is provided, take the first from the queue
            if self.booking is None:
                self.booking = heapq.heappop(simulation.Hotel.lobby_queue)

            # Calculate the service time and the time the service ends
            service_time = sampleCheckIn()
            service_end_time = self.time + service_time

            # Define the closing time for the lobby
            lobby_closing_time = 20 * 60  # 8:00 PM in minutes

            # If the current time exceeds the lobby closing time, release all rooms in the queue
            if self.time % 1440 > lobby_closing_time:
                while simulation.Hotel.lobby_queue:
                    booking = heapq.heappop(simulation.Hotel.lobby_queue)
                    simulation.Hotel.release_room(booking)
                return

            # Calculate the waiting time for this booking
            waitingTime = self.time - self.booking.arrivalTime

            # Decrease rank if the waiting time exceeds 20 minutes
            if waitingTime > 20:
                rankDecrease = (waitingTime // 20) * 0.02
                for customer in self.booking.customers:
                    customer.update_rank(rankDecrease)

            # Schedule the first day's pool event
            simulation.scheduleEvent(FirstDayPoolEvent(self.time, self.booking))

            # Calculate the checkout time based on the booking's stay duration
            start_day = int(self.booking.arrivalTime // 1440)
            checkout_time = (start_day + self.booking.stayDuration) * 1440 + 11 * 60  # 11:00 AM on the checkout day

            # Schedule a reminder for checkout
            simulation.scheduleEvent(CheckOutReminderEvent(checkout_time, self.booking))

            # If there are more bookings in the lobby queue, handle the next one
            if simulation.Hotel.lobby_queue:
                next_booking = heapq.heappop(simulation.Hotel.lobby_queue)
                simulation.scheduleEvent(CheckInDepartureEvent(service_end_time, next_booking))

class OpenBreakfast(Event):
    """
    Schedules the opening of the breakfast service and handles guest arrivals for breakfast.
    """
    def __init__(self, time):
        super().__init__(time)

    def handle(self, simulation):
        # Schedule the breakfast opening for the next day (24 hours later)
        simulation.scheduleEvent(OpenBreakfast(self.time + 24 * 60))

        # Retrieve the first booking from the breakfast list
        booking = simulation.Hotel.breakfast_list[0]

        # Calculate the arrival time for the booking based on the breakfast rate
        arrival_time = self.time + sampleBreakfastRate()

        # Schedule the breakfast arrival event for the calculated time
        simulation.scheduleEvent(BreakfastArrivalEvent(arrival_time, booking))

class CheckoutArrivalEvent(Event):
    """
    Handles the arrival of a guest for checkout. Assigns the guest to a server if available or adds them to the checkout queue.
    """
    def __init__(self, time, booking):
        super().__init__(time)
        self.booking = booking

    def handle(self, simulation):
        # If the guest has already checked out, do nothing
        if self.booking.hasCheckedOut is True:
            return

        # Mark the booking as checked out
        self.booking.hasCheckedOut = True

        # Attempt to find an available server for the checkout process
        server = simulation.Hotel.find_available_server()
        if server is not None:
            # Assign the server and schedule the checkout departure event
            simulation.Hotel.assign_to_server(server)
            service_time = self.time + sampleCheckOut()  # Calculate checkout service time
            simulation.scheduleEvent(CheckOutDepartureEvent(service_time, server, self.booking))
        else:
            # If no server is available, add the booking to the checkout queue
            simulation.Hotel.check_out_queue.append((self.time, self.booking))

class CheckOutDepartureEvent(Event):
    """
    Handles the completion of a guest's checkout process and updates hotel metrics.
    """
    def __init__(self, time, server_id, booking):
        super().__init__(time)
        self.server_id = server_id
        self.booking = booking

    def handle(self, simulation):
        # Determine the current day and the checkout time window (7:30 AM to 11:00 AM)
        current_day = self.time // 1440  # Convert time to days
        checkout_start = current_day * 1440 + 7.5 * 60  # Start time in minutes
        checkout_end = current_day * 1440 + 11 * 60  # End time in minutes

        # If the current time is past the checkout window, process all remaining bookings in the queue
        if self.time % 1440 > checkout_end:
            while simulation.Hotel.check_out_queue:
                booking = simulation.Hotel.check_out_queue.pop(0)
                feedback = self.booking.bookingRank()  # Collect feedback for the booking
                simulation.Hotel.hotelRatings.append(feedback)  # Update hotel ratings
                simulation.Hotel.dailyExpenses.append(self.booking.bookingCost(self.booking.groupType))  # Add to expenses
            return

        # Process the current booking: collect feedback and update expenses
        feedback = self.booking.bookingRank()
        simulation.Hotel.hotelRatings.append(feedback)
        simulation.Hotel.dailyExpenses.append(self.booking.bookingCost(self.booking.groupType))

        # If there are more bookings in the checkout queue, handle the next one
        if simulation.Hotel.check_out_queue:
            next_booking_time, next_booking = simulation.Hotel.check_out_queue.pop(0)
            service_time = self.time + sampleCheckOut()  # Calculate the checkout service time
            simulation.scheduleEvent(CheckOutDepartureEvent(service_time, self.server_id, next_booking))
        else:
            # Mark the server as free if no bookings are in the queue
            simulation.Hotel.server_states[self.server_id] = False

class BreakfastArrivalEvent(Event):
    """
    Handles the arrival of guests for breakfast, assigning them to tables if available or adding them to the breakfast queue.
    """
    def __init__(self, time, booking):
        super().__init__(time)
        self.booking = booking

    def handle(self, simulation):
        # Define breakfast service hours (6:30 AM to 11:30 AM)
        breakfast_start_time = 6.5 * 60
        breakfast_end_time = 11.5 * 60
        local_time = self.time % 1440

        # If the current time is after breakfast hours, process remaining bookings
        if local_time > breakfast_end_time:
            # Process the breakfast queue
            while simulation.Hotel.breakfast_queue:
                booking = simulation.Hotel.breakfast_queue.pop(0)
                # Schedule daily activities for bookings not on their last day
                if not booking.isLastDay(self.time // 1440 + 1):
                    simulation.scheduleEvent(DailyActivityEvent(self.time, booking))

            # Process any remaining breakfast bookings
            while simulation.Hotel.breakfast_list:
                booking = simulation.Hotel.breakfast_list.pop(0)
                if not booking.isLastDay(self.time // 1440 + 1):
                    simulation.scheduleEvent(DailyActivityEvent(self.time, booking))
            return

        # Record the breakfast arrival time for the booking
        self.booking.breakfastArrivalTime = self.time

        # Attempt to find an available table for the booking
        available_table_index = simulation.Hotel.find_available_table()
        if available_table_index is not None:
            # If a table is available, assign it and schedule a departure event
            breakfast_duration = self.time + sampleBreakfastTime()
            simulation.Hotel.tables[available_table_index].set_status(True)
            simulation.scheduleEvent(BreakfastDepartureEvent(breakfast_duration + 30, self.booking, available_table_index))
        else:
            # If no table is available, add the booking to the breakfast queue
            simulation.Hotel.breakfast_queue.append(self.booking)

        # If there are more bookings in the breakfast list, schedule their arrival
        if simulation.Hotel.breakfast_list:
            next_booking = simulation.Hotel.breakfast_list.pop(0)
            service_time = self.time + sampleBreakfastRate()
            simulation.scheduleEvent(BreakfastArrivalEvent(service_time, next_booking))

class BreakfastDepartureEvent(Event):
    """
    Handles the departure of a booking from the breakfast table, updates table status, and schedules the next steps for the booking.
    """
    def __init__(self, time, booking, table_id):
        super().__init__(time)
        self.booking = booking
        self.table_id = table_id

    def handle(self, simulation):
        # Mark the table as available
        simulation.Hotel.tables[self.table_id].set_status(False)

        # If it is not the last day for the booking, schedule daily activities
        if not self.booking.isLastDay(self.time // 1440 + 1):
            simulation.scheduleEvent(DailyActivityEvent(self.time, self.booking))

        # If it is the last day and past 7:30 AM, schedule a checkout arrival event
        local_time = self.time % 1440  # Convert to local time
        if self.booking.isLastDay(self.time // 1440 + 1) and local_time > 7.5 * 60:
            simulation.scheduleEvent(CheckoutArrivalEvent(self.time, self.booking))

        # Random chance to decrease customer rank due to dissatisfaction
        for customer in self.booking.customers:
            if random.random() < 0.1:  # 10% chance
                rankDecrease = 0.025
                customer.update_rank(rankDecrease)

        # If there are bookings in the breakfast queue, assign the next booking to the table
        if simulation.Hotel.breakfast_queue:
            next_booking = simulation.Hotel.breakfast_queue.pop(0)  # Get the next booking
            breakfast_duration = self.time + sampleBreakfastTime()  # Calculate breakfast duration
            simulation.scheduleEvent(BreakfastDepartureEvent(breakfast_duration, next_booking, self.table_id))

            # Calculate and store the wait time for the next booking
            wait_time = self.time - next_booking.breakfastArrivalTime
            simulation.Hotel.bookingsBreakfastWaitingTime.append(wait_time)

            # If the wait time exceeds 20 minutes, decrease customer rank
            if wait_time > 20:
                rankDecrease = (wait_time // 20) * 0.02
                for customer in self.booking.customers:
                    customer.update_rank(rankDecrease)

class CheckOutReminderEvent(Event):
    """
    Sends a reminder for the guest to check out. If the guest is in a breakfast-related queue or list, they are removed, and a checkout event is scheduled.
    """
    def __init__(self, time, booking):
        super().__init__(time)
        self.booking = booking

    def handle(self, simulation):
        # If the guest has not yet checked out
        if not self.booking.hasCheckedOut:
            # Remove the booking from the breakfast queue if it exists there
            if self.booking in simulation.Hotel.breakfast_queue:
                simulation.Hotel.breakfast_queue.remove(self.booking)

            # Remove the booking from the breakfast list if it exists there
            if self.booking in simulation.Hotel.breakfast_list:
                simulation.Hotel.breakfast_list.remove(self.booking)

            # Schedule a checkout arrival event for the guest
            simulation.scheduleEvent(CheckoutArrivalEvent(self.time, self.booking))

class FirstDayPoolEvent(Event):
    """
    Handles the scheduling of a booking's first pool activity. If the pool is available, the booking is added; otherwise, no action is taken.
    """
    def __init__(self, time, booking):
        super().__init__(time)
        self.booking = booking

    def handle(self, simulation):
        # Convert the event time to local time within the day (minutes since midnight)
        local_time = self.time % 1440

        # If the current time is after 7:00 PM, no pool activity is scheduled
        if local_time > 19 * 60:  # 7:00 PM in minutes
            return

        # Attempt to add the booking's group to the pool
        if simulation.Hotel.getFacility("pool").addCustomerToActivity(self.booking.getGroupSize()):
            # If successful, calculate the pool usage duration and schedule a departure event
            service_time = self.time + samplePoolSpent()
            simulation.scheduleEvent(FirstDayPoolDepartureEvent(service_time, self.booking))
        else:
            # If the pool is full or unavailable, no action is taken
            return

class FirstDayPoolDepartureEvent(Event):
    """
    Handles the departure of a booking from the pool after their first activity.
    """
    def __init__(self, time, booking):
        super().__init__(time)
        self.booking = booking

    def handle(self, simulation):
        # Mark the end of the pool activity for the booking's group
        simulation.Hotel.getFacility("pool").endCustomerActivity(self.booking.getGroupSize())

class DailyActivityEvent(Event):
    """
    Assigns daily activities (pool or bar) to customers based on their group type and probabilities.
    """
    def __init__(self, time, booking):
        super().__init__(time)
        self.booking = booking

    def handle(self, simulation):
        # Determine the type of group (family, couple, or individual)
        group_type = self.booking.checkGroupType()

        if group_type == "family":
            # For families, each customer has an 80% chance of going to the pool, 20% chance of going to the bar
            for customer in self.booking.customers:
                if np.random.rand() < 0.8:
                    simulation.scheduleEvent(PoolEvent(self.time, customer))
                else:
                    simulation.scheduleEvent(BarEvent(self.time, customer))

        elif group_type == "couple":
            # For couples, 70% chance both go to the pool, 30% chance only one goes
            if np.random.rand() < 0.7:
                for customer in self.booking.customers:
                    simulation.scheduleEvent(PoolEvent(self.time, customer))
            else:
                simulation.scheduleEvent(PoolEvent(self.time, self.booking.customers[0]))

        elif group_type == "individual":
            # Individuals always go to the pool
            for customer in self.booking.customers:
                simulation.scheduleEvent(PoolEvent(self.time, customer))

class PoolEvent(Event):
    """
    Handles a customer's pool activity, either by scheduling their pool time or adding them to the pool queue.
    """
    def __init__(self, time, customer):
        super().__init__(time)
        self.customer = customer

    def handle(self, simulation):
        # Get the pool facility
        facility = simulation.Hotel.getFacility("pool")
        local_time = self.time % 1440  # Convert the time to local time (minutes since midnight)

        # If the current time is after 7:00 PM, clear the pool and its queue
        if local_time > 19 * 60:  # 7:00 PM in minutes
            facility.currentOccupancy = 0
            facility.poolQueue = []
            return

        # If the current time is before 7:00 AM, schedule the pool activity for the next opening
        if local_time < 7 * 60:  # 7:00 AM in minutes
            current_day = self.time // (24 * 60)
            next_opening_time = (current_day * 24 * 60) + 7 * 60  # Next day at 7:00 AM
            simulation.scheduleEvent(PoolEvent(next_opening_time, self.customer))
            return

        # Update the customer's daily activity for the pool
        self.customer.daily_activities["pool"][1] += 1
        self.customer.last_activity = "pool"

        # Try to add the customer to the pool if there's available capacity
        if facility.addCustomerToActivity(1):
            service_time = self.time + samplePoolSpent()  # Calculate the pool duration
            simulation.scheduleEvent(PoolDepartureEvent(service_time, self.customer))  # Schedule departure
        else:
            # If the pool is full, add the customer to the pool queue
            facility.poolQueue.append(self.customer)
            facility.addToActivityQueue(self.customer)

            # Schedule a reneging event in case the customer leaves the queue
            simulation.scheduleEvent(RenegingEvent(self.time + self.customer.maxWaitingTime, self.customer, facility.poolQueue))

class BarEvent(Event):
    """
    Handles a customer's activity in the bar. If a barman is available, the customer is served; otherwise, the customer is added to the bar queue.
    """
    def __init__(self, time, customer):
        super().__init__(time)
        self.customer = customer

    def handle(self, simulation):
        # Update the customer's daily activity for the bar
        self.customer.daily_activities["bar"][1] += 1
        self.customer.last_activity = "bar"

        # Increment the total number of bar arrivals
        simulation.Hotel.bar_arrivals += 1

        # Check if a barman is available
        available_barman = simulation.Hotel.bar.findAvailableBarman()

        if available_barman is not None:
            # Assign the barman to the customer and schedule the bar departure event
            simulation.Hotel.bar.setBarmanStatus(available_barman, True)
            departure_time = self.time + sampleBarService()  # Calculate service time
            simulation.scheduleEvent(BarDepartureEvent(departure_time, self.customer, available_barman))
        else:
            # If no barman is available, add the customer to the bar queue
            simulation.Hotel.bar.barQueue.append(self.customer)

            # Schedule a reneging event in case the customer leaves the queue due to long wait
            simulation.scheduleEvent(RenegingEvent(self.time + self.customer.maxWaitingTime, self.customer, simulation.Hotel.bar.barQueue))

class SpaEvent(Event):
    """
    Handles a customer's activity in the spa. If space is available, the customer is admitted; otherwise, they are added to the spa queue.
    """
    def __init__(self, time, customer):
        super().__init__(time)
        self.customer = customer

    def handle(self, simulation):
        # Get the spa facility
        facility = simulation.Hotel.getFacility("spa")
        local_time = self.time % 1440  # Convert the event time to local time (minutes since midnight)

        # If the current time is after 7:00 PM, clear the spa and its occupancy
        if local_time > 19 * 60:  # 7:00 PM in minutes
            facility.currentOccupancy = 0
            return

        # If the current time is before 7:00 AM, schedule the spa event for the next opening
        if local_time < 7 * 60:  # 7:00 AM in minutes
            current_day = self.time // (24 * 60)
            next_opening_time = (current_day * 24 * 60) + 7 * 60  # Next day at 7:00 AM
            simulation.scheduleEvent(SpaEvent(next_opening_time, self.customer))
            return

        # Update the customer's daily activity for the spa
        self.customer.daily_activities["spa"][1] += 1
        self.customer.last_activity = "spa"

        # Attempt to add the customer to the spa if there's available capacity
        if facility.addCustomerToActivity(1):
            service_time = self.time + sampleSpa()  # Calculate the spa service duration
            simulation.scheduleEvent(SpaDepartureEvent(service_time, self.customer))  # Schedule spa departure
        else:
            # If the spa is full, add the customer to the spa queue
            facility.addToActivityQueue(self.customer)

            # Schedule a reneging event in case the customer leaves the queue due to long wait
            simulation.scheduleEvent(RenegingEvent(self.time + self.customer.maxWaitingTime, self.customer, facility.activityQueue))

class PoolDepartureEvent(Event):
    """
    Handles the departure of a customer from the pool and schedules their next activity or processes the next customer in the pool queue.
    """
    def __init__(self, time, customer):
        super().__init__(time)
        self.customer = customer

    def handle(self, simulation):
        # Get the pool facility
        facility = simulation.Hotel.getFacility("pool")

        # Mark the pool activity as completed for the customer
        self.customer.daily_activities["pool"][0] = True

        # Determine the customer's next activity after the pool
        next_activity = simulation.Hotel.getNextActivity(
            self.customer, self.customer.daily_activities, "pool"
        )

        # Schedule the next activity for the customer
        if next_activity == "bar":
            simulation.scheduleEvent(BarEvent(self.time, self.customer))
        elif next_activity == "spa":
            simulation.scheduleEvent(SpaEvent(self.time, self.customer))

        # End the customer's pool activity
        facility.endCustomerActivity(1)

        # If there are customers in the pool queue, process the next one
        if facility.poolQueue:
            next_customer = facility.poolQueue.pop(0)  # Get the next customer from the queue
            service_time = self.time + samplePoolSpent()  # Calculate the duration of their pool activity
            facility.addCustomerToActivity(1)  # Add the next customer to the pool
            simulation.scheduleEvent(PoolDepartureEvent(service_time, next_customer))  # Schedule their departure

class BarDepartureEvent(Event):
    """
    Handles the departure of a customer from the bar, updates their activities, and assigns the next customer in the queue to the barman if available.
    """
    def __init__(self, time, customer, barman):
        super().__init__(time)
        self.customer = customer
        self.barman = barman

    def handle(self, simulation):
        # Update the customer's bar expenses and mark the bar activity as completed
        self.customer.updateBarExpenses()
        self.customer.daily_activities["bar"][0] = True

        # Determine the customer's next activity after the bar
        next_activity = simulation.Hotel.getNextActivity(
            self.customer, self.customer.daily_activities, "bar"
        )

        # Schedule the next activity for the customer
        if next_activity == "pool":
            simulation.scheduleEvent(PoolEvent(self.time, self.customer))
        elif next_activity == "spa":
            simulation.scheduleEvent(SpaEvent(self.time, self.customer))

        # Check if there are customers waiting in the bar queue
        if simulation.Hotel.bar.barQueue:
            # Get the next customer from the queue
            next_customer = simulation.Hotel.bar.barQueue.pop(0)

            # Assign the barman to the next customer and schedule their service
            simulation.Hotel.bar.setBarmanStatus(self.barman, True)
            service_time = self.time + sampleBarService()  # Calculate the service duration
            simulation.scheduleEvent(BarDepartureEvent(service_time, next_customer, self.barman))
        else:
            # If no one is in the queue, mark the barman as available
            simulation.Hotel.bar.setBarmanStatus(self.barman, False)

class SpaDepartureEvent(Event):
    """
    Handles the departure of a customer from the spa, updates their activities, and assigns the next customer in the spa queue if available.
    """
    def __init__(self, time, customer):
        super().__init__(time)
        self.customer = customer

    def handle(self, simulation):
        # Get the spa facility
        facility = simulation.Hotel.getFacility("spa")

        # Mark the spa activity as completed and release the spa resources
        facility.endCustomerActivity(1)
        self.customer.daily_activities["spa"][0] = True

        # Determine the customer's next activity after the spa
        next_activity = simulation.Hotel.getNextActivity(
            self.customer, self.customer.daily_activities, "bar"
        )

        # Schedule the next activity for the customer
        if next_activity == "pool":
            simulation.scheduleEvent(PoolEvent(self.time, self.customer))
        elif next_activity == "bar":
            simulation.scheduleEvent(BarEvent(self.time, self.customer))

        # If there are customers waiting in the spa queue, assign the next one
        if facility.checkIfQueueIsNotEmpty():
            next_customer = facility.activityQueue.pop(0)  # Get the next customer from the queue
            service_time = self.time + sampleSpa()  # Calculate the spa service duration
            facility.addCustomerToActivity(1)  # Add the next customer to the spa
            simulation.scheduleEvent(SpaDepartureEvent(service_time, next_customer))  # Schedule their departure

class RenegingEvent(Event):
    """
    Handles the scenario where a customer leaves the queue due to excessive wait time (reneging).
    Updates their rank and schedules their next activity.
    """
    def __init__(self, time, customer, activity_queues):
        super().__init__(time)
        self.customer = customer
        self.activity_queues = activity_queues

    def handle(self, simulation):
        # Check if the customer is still in the activity queue
        if self.customer not in self.activity_queues:
            return

        # If the customer is leaving the bar queue, update the bar's reneging count
        if self.customer.last_activity == "bar":
            simulation.Hotel.bar_reneging_events += 1

        # Remove the customer from the current activity queue
        self.activity_queues.remove(self.customer)

        # Decrease the customer's rank due to the reneging event
        self.customer.update_rank(0.03)

        # Determine the customer's next activity after leaving the queue
        next_activity = simulation.Hotel.getNextActivity(
            self.customer, self.customer.daily_activities, self.customer.last_activity
        )

        # Schedule the next activity for the customer
        if next_activity == "bar":
            simulation.scheduleEvent(BarEvent(self.time, self.customer))
        elif next_activity == "pool":
            simulation.scheduleEvent(PoolEvent(self.time, self.customer))
        elif next_activity == "spa":
            simulation.scheduleEvent(SpaEvent(self.time, self.customer))


class Simulation:
    def __init__(self, simulationTime, numberOfTables,numberOfBarmens):
        self.simulationTime = simulationTime
        self.Hotel = Hotel(numberOfTables,numberOfBarmens)
        self.clock = 0
        self.event_list = []
        self.total_waiting_time = 0

    def scheduleEvent(self, event):
        heapq.heappush(self.event_list, event)


    def run(self):
        self.scheduleEvent(OpenHotelEvent(480))
        self.scheduleEvent(OpenBreakfast(1830))#first meal
        self.scheduleEvent(closeHotelDayEvent(24*60))

        while self.event_list and self.clock < self.simulationTime:
            event = heapq.heappop(self.event_list)
            self.clock = event.time
            event.handle(self)
```

  <div style="direction:rtl" dir="rtl">
שיטת וולש (Welch method) היא טכניקת חישוב ממוצעים נעים (Moving Average) שנועדה להחליק את הנתונים ולהפחית רעש (Noise) על ידי חישוב ממוצע של קבוצות נתונים לפי חלון גודל משתנה.
במהלך הסימולציה, השתמשנו בשיטת וולש על מספר מדדים שמטרתנו הייתה לעקוב אחריהם לאורך הזמן ולבחון את השפעתם:
  <div style="direction:rtl" dir="rtl">
1. דירוג יומי ממוצע של המלון
  <div style="direction:rtl" dir="rtl">
2. זמן המתנה ממוצע לארוחת הבוקר
  <div style="direction:rtl" dir="rtl">
3. אחוזי הנטישה של התור בבר

  <div style="direction:rtl" dir="rtl">
  בהתאם לכך, השתמשנו בשיטת וולש כדי להפחית רעש וליצור גרפים חלקים יותר עבור המדדים שבחנו בסימולציה. גודל החלון לקביעת הממוצעים נקבע מראש, אך יכול להשתנות לפי הצורך, כאשר מטרתנו הייתה למצוא את נקודת הזמן שבה המדדים מגיעים ליציבות, או מה שמכונה "זמן החימום". לשם כך, חישבנו את הממוצעים לכל חלון, והגדלנו את גודל החלון במידת הצורך כדי להגיע ליציבות טובה יותר, ובסופו של דבר, להגדיר את זמן החימום כנקודת הזמן בה המדד מייצב את עצמו.


```python
import numpy as np
import matplotlib.pyplot as plt

def welch_method(data, w):
    n = len(data)
    welch_averages = []

    for i in range(n):
        if i < w:
            start_idx = 0
            end_idx = i + i + 1
            numerator = np.sum(data[start_idx:end_idx])
            denominator = 2 * (i + 1) - 1
        elif i < n - w:
            start_idx = i - w
            end_idx = i + w + 1
            numerator = np.sum(data[start_idx:end_idx])
            denominator = 2 * w + 1
        else:
            start_idx = max(0, i - w)
            end_idx = n
            numerator = np.sum(data[start_idx:end_idx])
            denominator = end_idx - start_idx
        avg = numerator / denominator
        welch_averages.append(avg)

    return welch_averages

def heatingTime(simulationTime,numberOfTables, numberOfBarmens, wRank, wBreakfast, wabandonmen, label):
    s = Simulation(simulationTime*60*24, numberOfTables,numberOfBarmens)
    s.run()
    welsh_dailyRank = welch_method(s.Hotel.dailyHotelRanking, wRank)
    plt.figure(figsize=(10, 6))
    labelUnited= label +" Daily Rank"
    plt.plot(range(1, len(s.Hotel.dailyHotelRanking) + 1), s.Hotel.dailyHotelRanking, label=labelUnited, color='blue')
    plt.plot(range(1, len(s.Hotel.dailyHotelRanking) + 1), welsh_dailyRank, label='Welsh Averages (Adjusted at End)', linestyle='--', color='red')
    plt.title(labelUnited)
    plt.xlabel("Day")
    plt.ylabel("Rank")
    plt.legend()
    plt.show()

    welsh_dailyBreakfastWaitingTime = welch_method(s.Hotel.dailyBreakfastWaitingTime, wBreakfast)
    plt.figure(figsize=(10, 6))
    plt.ylim(0, 60)
    labelUnited=label+" Daily Waiting For Breakfast Time"
    plt.plot(range(1, len(s.Hotel.dailyBreakfastWaitingTime) + 1), s.Hotel.dailyBreakfastWaitingTime, label=labelUnited, color='blue')
    plt.plot(range(1, len(s.Hotel.dailyBreakfastWaitingTime) + 1), welsh_dailyBreakfastWaitingTime, label='Welsh Averages (Adjusted at End)', linestyle='--', color='red')
    plt.title(labelUnited)
    plt.xlabel("Day")
    plt.ylabel("Waiting For Breakfast Time")
    plt.legend()
    plt.show()


    welsh_barabandonment = welch_method(s.Hotel.daily_bar_abandonment_rates, wabandonmen)
    plt.figure(figsize=(10, 6))
    labelUnited=label+" Daily Bar abandonment"
    plt.plot(range(1, len(s.Hotel.daily_bar_abandonment_rates) + 1), s.Hotel.daily_bar_abandonment_rates, label=labelUnited, color='blue')
    plt.plot(range(1, len(s.Hotel.daily_bar_abandonment_rates) + 1), welsh_barabandonment, label='Welsh Averages (Adjusted at End)', linestyle='--', color='red')
    plt.title(labelUnited)
    plt.xlabel("Day")
    plt.ylabel("Daily abandonment in %")
    plt.legend()
    plt.show()
    return s

originalSimulation = heatingTime(150, 12,2, 3, 37, 37,"Baseline simulation")
```

```python
import statistics
import numpy as np
from scipy import stats
import pandas as pd

baselineDailyHotelRank=[]
baselineBreakfastWaitingTime = []
baselineDailyHotelIncomes = []
baselineAbandonedBar = []
alternativeDailyHotelRank=[]
alternativeBreakfastWaitingTime = []
alternativeDailyHotelIncomes = []
alternativeAbandonedBar = []

for i in range (15):

    baselineSimulation = Simulation(525*60*24,12,2)
    baselineSimulation.run()
    alternativeSimulation = Simulation(700*60*24,16,3)
    alternativeSimulation.run()

    baselineDailyHotelRank.append(statistics.mean(baselineSimulation.Hotel.dailyHotelRanking[75:525]))
    baselineBreakfastWaitingTime.append(statistics.mean(baselineSimulation.Hotel.dailyBreakfastWaitingTime[75:525]))
    baselineAbandonedBar.append(statistics.mean(baselineSimulation.Hotel.daily_bar_abandonment_rates[75:525]))

    alternativeDailyHotelRank.append(statistics.mean(alternativeSimulation.Hotel.dailyHotelRanking[100:700]))
    alternativeBreakfastWaitingTime.append(statistics.mean(alternativeSimulation.Hotel.dailyBreakfastWaitingTime[100:700]))
    alternativeAbandonedBar.append(statistics.mean(alternativeSimulation.Hotel.daily_bar_abandonment_rates[100:700]))

# Calculate relative accuracy for each metric
def calculate_relative_accuracy(baseline, alternative, relative_accuracy=0.1, confidence_level=0.9):
  n = 15
  baseline_mean = np.mean(baseline)
  baseline_std = np.std(baseline, ddof=1)
  alternative_mean = np.mean(alternative)
  alternative_std = np.std(alternative, ddof=1)
  alfa_adj = 0.1/6

  # t-value for the given confidence level
  t_critical = stats.t.ppf(1 - alfa_adj/2 , df=n - 1)

  # Calculate delta and relative accuracy
  delta_baseline = t_critical * (baseline_std / np.sqrt(n))
  delta_alternative = t_critical * (alternative_std / np.sqrt(n))
  relative_accuracy_baseline = delta_baseline / baseline_mean
  relative_accuracy_alternative = delta_alternative / alternative_mean

  return {
      "Baseline Mean": baseline_mean,
      "Alternative Mean": alternative_mean,
      "Delta Baseline": delta_baseline,
      "Delta Alternative": delta_alternative,
      "Relative Accuracy Baseline": relative_accuracy_baseline,
      "Relative Accuracy Alternative": relative_accuracy_alternative,
      "Meets Accuracy (Baseline)": relative_accuracy_baseline <= relative_accuracy,
      "Meets Accuracy (Alternative)": relative_accuracy_alternative <= relative_accuracy
  }

# Metrics data
metrics = {
  "Daily Hotel Rank": (baselineDailyHotelRank, alternativeDailyHotelRank),
  "Breakfast Waiting Time": (baselineBreakfastWaitingTime, alternativeBreakfastWaitingTime),
  "Abandoned Bar Rate": (baselineAbandonedBar, alternativeAbandonedBar)
}

# Store results in a DataFrame (assuming you've populated the data)
results = []
for metric_name, (baseline, alternative) in metrics.items():
  result = calculate_relative_accuracy(baseline, alternative)
  result["Metric"] = metric_name
  results.append(result)
df_results = pd.DataFrame(results)

# Print the results as a string
print(df_results.to_string())

results_daily_rank = calculate_relative_accuracy(
    baseline=baselineDailyHotelRank,
    alternative=alternativeDailyHotelRank,
    relative_accuracy=0.1,
    confidence_level=0.9
)

results_breakfast_waiting_time = calculate_relative_accuracy(
    baseline=baselineBreakfastWaitingTime,
    alternative=alternativeBreakfastWaitingTime,
    relative_accuracy=0.1,
    confidence_level=0.9
)

results_abandoned_bar = calculate_relative_accuracy(
    baseline=baselineAbandonedBar,
    alternative=alternativeAbandonedBar,
    relative_accuracy=0.1,
    confidence_level=0.9
)
```

 לאחר בחינת החלופה והמצב הקיים, נרצה לבחור את החלופה הטובה ביותר בהתבסס על המדדים שנבחרו.
    לשם כך, השתמשנו בשיטת **Welch**, המאפשרת השוואה מדויקת בין שתי קבוצות נתונים שאינן בהכרח בעלות שוויון שונויות.

    הניתוח כלל גם חישוב t-statistic עבור רמת מובהקות של 10%, וחישוב רווחי סמך באמצעות t-קריטי.
    תהליך זה אפשר לנו להעריך איזו חלופה עדיפה עבור כל מדד שנבחר.

**ההנחות של שיטה זו:**
1. אין תלות בין ריצות מקבילות בשתי הסדרות.
2. אין תלות בין ריצות בכל חלופה בפני עצמה.
3. לא ניתן להניח שוויון שונויות בין הסדרות.
4. התוצאות מתפלגות נורמאלי.


היות והשתמשנו בשיטת Replication אין תלות בין ריצות מקבילות בשתי הסדרות ואין תלות בין ריצות בכל חלופה בפני עצמה. בנוסף, אין מספיק נתונים על מנת להניח שוויון שונויות ועל פי משפט הגבול המרכזי נוכל להסיק כי התוצאות מתפלגות נורמאלי.
שיטה זו משתמשת בהפרש של ממוצעי המדדים עבור החלופה והמצב הקיים במקרה שלנו ובעזרתו יוצרת רווח סמך.

**בעזרת רווח הסמך נוכל להסיק איזו חלופה עדיפה:**
1. רווח הסמך חיובי-המצב הקיים גדול מהאלטרנטיבה.
2. רווח הסמך שלילי- האלטרנטיבה גדולה מהמצב הקיים.
3. רווח הסמך כולל את 0- לא נוכל להבדיל בינהן.



  <p>

```python
import pandas as pd
import numpy as np
from scipy import stats

def calculate_confidence_interval(data1, data2, alpha=0.1):
    """
    Calculate the confidence interval for the difference in means between two datasets.
    """
    n1, n2 = 15, 15
    mean1, mean2 = np.mean(data1), np.mean(data2)
    var1, var2 = np.var(data1, ddof=1), np.var(data2, ddof=1)

    # Difference in means
    mean_diff = mean1 - mean2

    # Calculate the standard error of the difference
    standard_error = np.sqrt((var1 / n1) + (var2 / n2))

    # Degrees of freedom
    numerator = (var1 / n1 + var2 / n2) ** 2
    denominator = ((var1**2) / ((n1**2) * (n1 - 1))) + ((var2**2) / ((n2**2) * (n2 - 1)))
    degrees_of_freedom = int(numerator / denominator)

    # Critical t-value using adjusted alpha
    adjusted_alpha = alpha / 6
    t_crit = stats.t.ppf(1 - adjusted_alpha / 2, degrees_of_freedom)

    # Confidence interval
    margin_of_error = t_crit * standard_error
    ci_lower = mean_diff - margin_of_error
    ci_upper = mean_diff + margin_of_error

    # Return the results
    return mean_diff, (ci_lower, ci_upper), degrees_of_freedom, t_crit


metrics = {
    "Daily Hotel Rank": (baselineDailyHotelRank, alternativeDailyHotelRank),
    "Breakfast Waiting Time": (baselineBreakfastWaitingTime, alternativeBreakfastWaitingTime),
    "Abandoned Bar Rate": (baselineAbandonedBar, alternativeAbandonedBar)
}

alpha = 0.1
results = []

for metric_name, (baseline_data, alternative_data) in metrics.items():
    if not baseline_data or not alternative_data:  # Check if data is empty
        print(f"{metric_name}: Insufficient data to calculate confidence interval.")
        continue

    mean_diff, (ci_lower, ci_upper), degrees_of_freedom, t_crit = calculate_confidence_interval(
        baseline_data, alternative_data, alpha
    )
    results.append([metric_name, mean_diff, ci_lower, ci_upper, degrees_of_freedom])

    # Output results
    print(f"{metric_name}:")
    print("  Difference in means:", mean_diff)
    print("  Degrees of Freedom (approx):", degrees_of_freedom)
    print("  Critical t-value:", t_crit)
    print("  Confidence Interval: ({:.4f}, {:.4f})".format(ci_lower, ci_upper))

    # Interpretation
    if ci_upper < 0:
        print("  Preferred Alternative: The second in the pair (negative difference).")
    elif ci_lower > 0:
        print("  Preferred Alternative: The first in the pair (positive difference).")
    else:
        print("  No significant difference between the alternatives.")
    print()
```
